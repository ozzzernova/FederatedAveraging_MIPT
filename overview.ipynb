# **Обзор статьи:**

**Проблема и примеры возникновения:**

Современные мобильные устройства имеют доступ к обширным данным, пригодным для обучения моделей, что в свою очередь может значительно улучшить пользовательский опыт на устройстве. Например, языковые модели могут улучшить распознавание речи и ввод текста, а модели изображений могут автоматически выбирать хорошие фотографии. Однако эти богатые данные часто являются конфиденциальными, имеют большой объем или и то, и другое, что может исключить возможность регистрации в центре обработки данных и обучения там с использованием традиционных методов.

\
**Решение и идея статьи:**

Предлагается подход, который предполагает оставление распределенных данных для обучения на мобильных устройствах и обучение общей модели путем агрегирования локальных вычисленных обновлений. Это называется Федеративным обучением. У каждого участвующего устройства(клиента) есть локальный обучающий набор данных, который никогда не загружается на сервер. Вместо этого каждый клиент вычисляет обновление текущей глобальной модели, поддерживаемой сервером, и только это обновление передается. В статье рассматривается алгоритм FederatedAveraging, объединяющий локальный стохастический градиентный спуск(GCD), выполняющийся на каждом клиенте, и сервер, который усредняет полученные модели. Все это направлено на устойчивость к несбалансированным и не-идентично и не-независимо распределенным данным и сокращение количество раундом обмена данными.

\
**Постановка задачи:**

Сначала рассмотрим свойства задач федеративной оптимизации:

- **Уникальные данные:** Обучающие данные на конкретном клиенте обычно основаны на использовании мобильного устройства конкретным пользователем, и, следовательно, локальный набор данных конкретного пользователя не будет представительным для распределения в популяции.
- **Разный объем данных:** Точно так же некоторые пользователи будут гораздо активнее использовать сервис или приложение, чем другие, что приводит к различным объемам локальных обучающих данных.
- **Большое число клиентов:** Мы ожидаем, что количество клиентов, участвующих в оптимизации, будет значительно превышать среднее количество примеров на клиенте.
- **Разрозненность и ограниченные коммуникации** Мобильные устройства часто находятся в автономном режиме или используют медленные или дорогостоящие соединения.

Рассматриваем минимизацию конечных сумм

$$ \min\limits_{w \in \mathbb{R}^d} f(w), \ \ f(w) := \frac{1}{n} \sum\limits_{i = 1}^n f_i(w) $$

Для задач машинного обучениея в качестве $f_i$ берем функцию потерь $ l(x_i, y_i, w) $.Предполагается, что данные распределены на $K$ клиентов с разбиением $\mathcal{P}_k, \ n_k = | \mathcal{P}_k |$. Тогда функция принимает вид:

$$ f(w) = \sum\limits_{k = 1}^K \frac{n_k}{n}F_k(w), \ \ F_k = \frac{1}{n_k} \sum\limits_{i \in \mathcal{P}_k} f_i(w) $$

_Так же идеальная задача для федеративного обучения обладает следующими свойствами:_

- Обучение на реальных данных с мобильных устройств предоставляет определенные преимущества по сравнению с обучением на прокси-данных, которые обычно доступны в центре обработки данных.

- Эти данные чувствительны к конфиденциальности или обладают большим размером (по сравнению с размером модели), поэтому предпочтительно не записывать их в центр данных исключительно в целях обучения модели (в соответствии с принципом фокусированного сбора

- Для задач с учителем метки на данных могут быть естественно выведены из пользовательского взаимодействия.
